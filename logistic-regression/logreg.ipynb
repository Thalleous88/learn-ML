### **Binary Logistic Regression**

```python
import numpy as np
```

```python
# Features: [Hours Studied, Practice Exams Taken]
X = [
    [1.5, 0], [2.0, 1], [3.5, 0], [2.5, 2], 
    [4.0, 1], [3.0, 0], [1.0, 1], [4.5, 2], # Students who Failed
    [5.0, 3], [6.5, 4], [7.0, 3], [5.5, 5], 
    [8.0, 4], [6.0, 3], [9.0, 5], [7.5, 4]  # Students who Passed
]

y = [
    0, 0, 0, 0, 0, 0, 0, 0, # Fails
    1, 1, 1, 1, 1, 1, 1, 1  # Passes
]
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def bcel(ytrue, ypred):
    return -1 * (np.dot(ytrue, np.log(sigmoid(ypred))) + np.dot((1-ytrue), np.log(sigmoid(1-ypred))))
```

```python
class logreg:
    def __init__(self):
        self.weight = np.array([1.0, 1.0])
        self.bias = 0.0
        
    def predict(self, x):
        out = self.weight @ x + self.bias
        res = sigmoid(out)
        
        if res >= 0.5:
            return "Pass"
        else:
            return "Fail"
        
    def forward_pass(self, x):
        out = self.weight @ x + self.bias
        return sigmoid(out)
    
    def backprop(self, X, y):
        lr = 0.01
        epochs = 1000
        
        for i in range(epochs):
            
            total_dL_dw = np.zeros_like(self.weight)
            total_dl_db = 0.0
            epoch_loss = 0.0
            
            
            ypred = [self.forward_pass(x) for x in X]
            for idx, (yp, ytrue) in enumerate(zip(ypred, y)):
            # binary cross entropy
            # L(x) = y * log(sigmoid(x)) + (1-y) * log(1-sigmoid(x))
            # u = sigmoid(x)
            # dL/du = y/u + (1-y)/(1-u)
            # du/dout = sigmoid(x) * (1-sigmoid(x))
            # dout/dw = x  
            # dout/db = 1
            
            
            

                dL_dypred = (-1 * ytrue/yp) + (1-ytrue)/(1-yp)
                dypred_dout = yp * (1-yp)
                dL_dout = dL_dypred * dypred_dout
                
                
                total_dL_dw += np.dot(dL_dout, X[idx])
                total_dl_db += dL_dout 
                
                epoch_loss += bcel(ytrue, yp)
                
               
                
                
            self.weight -= float(lr) * total_dL_dw        
            self.bias -= float(lr) * total_dl_db
            
            print(f"Epoch {i+1}/{epochs}, Loss: {epoch_loss/len(X)}")
                
    
    
```

```python
instance = logreg()

instance.backprop(X, y)
```

```python
instance.predict([1, 2])
```

### **Multivariate Logistic Regression**

```python
import numpy as np

# Features: [Study Hours per Week, Attendance Percentage]
X_grades_list = [
    # Grade 'F'
    [0.5, 10], [1.0, 25], [2.0, 15], [0.0, 5], [1.5, 30],
    
    # Grade 'D'
    [3.0, 50], [2.5, 60], [4.0, 45], [3.5, 55], [2.0, 65],
    
    # Grade 'C'
    [5.0, 75], [6.5, 70], [5.5, 80], [7.0, 65], [6.0, 75],
    
    # Grade 'B'
    [9.0, 90], [8.5, 85], [10.0, 80], [8.0, 95], [9.5, 88],
    
    # Grade 'A'
    [14.0, 98], [12.5, 100], [15.0, 95], [13.0, 99], [11.5, 97]
]

y_grades_list = [
    'F', 'F', 'F', 'F', 'F',
    'D', 'D', 'D', 'D', 'D',
    'C', 'C', 'C', 'C', 'C',
    'B', 'B', 'B', 'B', 'B',
    'A', 'A', 'A', 'A', 'A'
]

X_grades = np.array(X_grades_list)
y_grades = np.array(y_grades_list)
```

```python
def softmax(input):
    input = input - max(input)
    tot = sum([np.exp(z) for z in input])
    
    res = np.array([np.exp(z) / tot for z in input])
    
    return res

def ccel(ytrue, ypred):
    ypred = np.clip(ypred, 1e-15, 1.0)
    return -np.sum(ytrue * np.log(ypred))
```

```python
class mvlogreg:
    def __init__(self, input, target):
        feature_dim = input.shape[1]
        target_dim = len(np.unique(target))
        self.weight = np.ones((target_dim, feature_dim)) * 0.01
        self.bias = np.zeros((target_dim))
        
    
    def one_hot_encoding(self, target):
        hashmap = {}
        new = np.zeros(len(target), dtype=int)
        
        if isinstance(target[0], (np.str_, str)):
            idx = 0
            for i, category in enumerate(target):
                if category not in hashmap:
                    hashmap[category] = idx
                    idx += 1
                
                new[i] = hashmap[category]
            
        num_classes = self.bias.shape[0]
        res = np.eye(num_classes)[new]
        
        return res, hashmap
    

    def forward_pass(self, input):
        out = self.weight @ input + self.bias
        
        return softmax(out)
    
    
    def predict(self, input):
        out = self.weight @ input + self.bias
        out = softmax(out)
        
        return np.argmax(out)
    
    
    def backprop(self, input, target, lr=0.001, epochs=3000):
        for epoch in range(epochs):
            ypred = [self.forward_pass(i) for i in input]
            total_dL_dw = np.zeros_like(self.weight)
            total_dl_db = np.zeros_like(self.bias)
            epoch_loss = 0
            
            for idx, (yp, ytrue) in enumerate(zip(ypred, target)):
                dL_dout = yp - ytrue
                total_dL_dw += np.outer(dL_dout, input[idx])
                total_dl_db += dL_dout 
                
                epoch_loss += ccel(ytrue, yp)
                
            self.weight -= lr * total_dL_dw / len(input)
            self.bias -= lr * total_dl_db / len(input)
            
            if (epoch + 1) % 100 == 0:
                print(f"Epoch {epoch+1}/{epochs}, Loss: {epoch_loss/len(input):.4f}")
            
```

```python
model = mvlogreg(X_grades, y_grades)

y_encoded, labels = model.one_hot_encoding(y_grades)

labels = dict(zip(labels.values(), labels.keys()))


```

```python
model.backprop(X_grades, y_encoded)
```

```python
print(labels[model.predict(X_grades[0])])
```

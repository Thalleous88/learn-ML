```python
import numpy as np
import json
```

```python
with open('dataset.json', 'r') as file:
    data = json.load(file)
```

```python
def create_count_table(data, target_label):
    count = {}

    for item in data.values():
        for col in item.keys():
            if (col in count and item[col] in count[col]) and item[target_label] in count[col][item[col]]:
                count[col][item[col]][item[target_label]] += 1
            else:
                if col not in count:
                    count[col] = {}
                
                if item[col] not in count[col]:
                    count[col][item[col]] = {}
                    
                count[col][item[col]][item[target_label]] = 1
                
    return count
    
def get_target_classes(data, target_label):
    target_class = []

    for item in data.values():
        for col in item.keys():
            if col == target_label and item[col] not in target_class:
                target_class.append(item[col])
                
    return target_class
```

```python
class dtree:
    def __init__(self, data, target_label):
        self.count_table = create_count_table(data, target_label)
        self.target_class = get_target_classes(data, target_label)
        self.row_count = len(data)
        
        
        
    def entropy(self, label, cls):
        tot = sum([self.count_table[label][cls][t] for t in self.target_class if t in self.count_table[label][cls]])
        
        return -1 * sum([(self.count_table[label][cls][t]/tot) * np.log2(self.count_table[label][cls][t]/tot) for t in self.target_class if t in self.count_table[label][cls]])
    
    def information_gain(self, label):
        classes = get_target_classes(data, label)
        class_count_table = {}
        
        for cls in classes:
            for t in self.target_class:
                if t in self.count_table[label][cls]:
                    if t in class_count_table:
                        class_count_table[t] += self.count_table[label][cls][t]
                    else:
                        class_count_table[t] = self.count_table[label][cls][t]
                    
                
        total_entropy = -1 * sum([i/self.row_count * np.log2(i/self.row_count) for i in class_count_table.values()])
        
        return total_entropy - sum([sum(self.count_table[label][cls].values())/self.row_count * self.entropy(label, cls) for cls in classes])
```

```python
model = dtree(data, 'action')

model.count_table['device_type'].keys()
```

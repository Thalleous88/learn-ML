```python
import numpy as np
```

```python
X = np.array([
    [150, 2],
    [170, 3],
    [160, 1],
    [120, 8],
    [110, 9],
    [130, 7]
])


y = np.array(["Orange", "Orange", "Orange", "Lemon", "Lemon", "Lemon"])
```

```python
class knn:
    def euclidean_distance(self, data1, data2):
        return np.sqrt(np.sum(np.square(data1-data2)))
    
    def predict(self, input, data, label, k):
        arr = np.array([self.euclidean_distance(input, i) for i in data])
        
        label = label.copy()
        for i in range(len(data)):
            for j in range(len(data)-1):
                if arr[j] > arr[j+1]:
                    arr[j+1], arr[j] = arr[j], arr[j+1]
                    label[j+1], label[j] = label[j], label[j+1]
        
        counter = {}
        for i in range(k):
            if label[i] not in counter:
                counter[label[i]] = 0
            counter[label[i]] += 1
        
        res = max(counter, key=counter.get)
        
        return res
        
        
            
        
        
```

```python
model = knn()

model.predict(np.array([140, 6]), X, y, 4)
```
